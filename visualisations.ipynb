```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
df
```

```python
plt.scatter(x=df.age, y=df.thalach)
plt.xlabel("Age")
plt.ylabel("Thalach")
plt.title("Age Vs Thalach")
plt.show()
```

```python
plt.scatter(x=df.age, y=df.chol)
plt.xlabel("Age")
plt.ylabel("Cholestrol")
plt.title("Age Vs Cholestrol")
plt.show()
```

```python
plt.scatter(x=df.age, y=df.chol, c="red", marker='*')
plt.scatter(x=df.age, y=df.trestbps, c="blue", marker='x')
plt.scatter(x=df.age, y=df.thalach, c="green", marker='+')
plt.legend(['Chol','Trst-bps','Thalach'])
plt.show()
```

```python
count_df=df.copy()
count_df.sex = count_df.sex.apply(lambda x:"Male" if x==1 else "Female")
sns.histplot(x=count_df.sex)
```

```python
sns.countplot(x=df.age, hue=count_df.sex)
plt.xticks(label=df.age, rotation=90)
plt.show()
```
