```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dataset=pd.read_csv('dataset.csv')
dataset
```

```python
#linear regression
x = np.array(dataset.depth)
y = np.array(dataset.age)
n = np.size(x)
#mean of x and y
mean_x = np.sum(x)/n
mean_y = np.sum(y)/n
#summation values taken
summation_x_y = np.sum(y*x) - n*mean_y*mean_x
summation_x_x = np.sum(x*x) - n*mean_x*mean_x
b_1 = summation_x_y /summation_x_x
b_0 = mean_y - b_1*mean_x
print("Estimated coefficients:\nb_0 =",b_0," \nb_1 =" ,b_1)
plt.scatter(x, y,s = 30)
y_predicted = b_0 + b_1*x
plt.plot(x, y_predicted)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
